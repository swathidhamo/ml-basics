```python
from scipy.stats import multivariate_normal
from sklearn.cross_validation import train_test_split
mean_1 = []
mean_2 = []
std = []
covariance = []
for i in range(0, 20):
    mean_1.append(random.uniform(1, 10))
    mean_2.append(random.uniform(1, 10))
    std.append(random.uniform(1, 10))

for i in range(0, 20):
    temp = []
    for j in range(0, 20):
        if i==j:
            temp.append(std[i])
        else:
            k = random.uniform(0.1, 10)
            if k < 0:
                k = k * -1
            temp.append(k)
    covariance.append(temp)
covariance
# a matrix of random values shaped (1000,1000)
draw1 = np.random.multivariate_normal(mean_1, covariance,  10)
draw2 = np.random.multivariate_normal(mean_2, covariance,  10)
X_train, X_test = train_test_split(draw1, test_size=0.3)
Y_train, Y_test = train_test_split(draw2, test_size=0.3)
#plt.scatter(draw[:1,'o',c='green');
#plt.scatter(nexte,'o',c='blue')
#plt.axis('equal');
#print "First dist", draw, "\n", "Second dist", draw2
#plt.show()
```
